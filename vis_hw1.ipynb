```python
import networkx as nx
import pylab as plt
```

```python
tree = nx.read_graphml('tree-60n.xml')
```

```python
tree_dict = {}
for node in tree.nodes:
    tree_dict[node] = list(tree.neighbors(node))
```

```python
def HVtree(root, depth, width):
    positions[root] = (width, depth)
    children = tree_dict[root]
    if len(children) == 0:
        return 0
    elif len(children) == 1:
        return HVtree(children[0], depth, width+1)
    else:
        depth_right = HVtree(children[0], depth, width+1)
        depth_down = HVtree(children[1], depth+depth_right-1, width)
        return depth_right+depth_down-1
    
```

```python
positions = {}
HVtree(list(tree.nodes)[0], 0, 0)
```

```python
plt.figure(figsize=(20, 20))
nx.draw(tree, positions, with_labels=True, arrowsize=25, node_size=800, node_color = '#00ff00')
```
